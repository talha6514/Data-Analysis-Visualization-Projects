## Iris Dataset

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris) # Plots pairwise relationship among different attributes
```

### Pairplot does everything automatically but PairGrid can give more control over it.

```python
sns.PairGrid(iris)
```

```python
g = sns.PairGrid(iris) # We can call matplotlib methods using PairGrid for mapping or plotting
g.map(plt.scatter)
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

## Using FacetGrid (Lets go back to Tips dataset)

### Size & Aspect

We can use matplotlib's figsize parameter to change the size and aspect of the figures through seaborn

```python
plt.figure(figsize = (12,3))
sns.countplot(x = 'sex', data = tips)
```

## Scale & Context

set_context() allows you to override default parameters and change font scales.

```python
sns.set_context('poster') # We can add another parameter for fonts "font_scale = 1"
sns.countplot(x = 'sex', data = tips)
```
