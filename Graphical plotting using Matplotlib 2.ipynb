```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
```

```python
x
```

```python
y = x**2
```

```python
y
```

## 1) Functional Method 

```python
plt.plot(x,y)
```

```python
# if you want a coloured graph other than the default colour
plt.plot(x,y,'r-')
```

```python
# adding labels to the graph
plt.plot(x,y,'r-')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
# multiple plots on the same canvas
plt.subplot(1,2,1)
plt.plot(x,y,'r-')
plt.title('Title1')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
plt.title('Ttitle2')
```

# 2) Object Oriented Method

First we will make objects and then we will call methods from that objets.

```python
fig = plt.figure() #fig is an object which is a blank canvas
axes = fig.add_axes([0.1,0.1,0.8,0.8]) #add_axes([left, bottom, width, height])
axes.plot(x,y)
```

```python
# setting labels
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Titile')
```

```python
# two figures on one canvas OOP
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) #add_axes([left, bottom, width, height])
axes2 = fig.add_axes([0.2,0.4,0.3,0.4])
axes1.plot(x,y)
axes1.set_title('Title 1')
axes2.plot(y,x)
axes2.set_title('Title 2')
```

```python
# subplots using OOP
fig,axes = plt.subplots(nrows = 1,ncols = 2) #working as axes.plot() and as well as plt.figure()
for current_ax in axes:   #axes is list of the axes of all the subplots
    current_ax.plot(x,y,'g')
plt.tight_layout()
```

```python
#since iteration is possible so indexing is also possible
fig,axes = plt.subplots(nrows = 1,ncols = 2)
axes[0].plot(x,y)
axes[0].set_title('Title1')
axes[0].set_xlabel('X Label')
axes[0].set_ylabel('Y Label')
axes[1].plot(y,x)
axes[1].set_title('Title2')
axes[1].set_xlabel('X Label')
axes[1].set_ylabel('Y Label')
plt.tight_layout()
```

# Figure size, aspect ration and DPI

```python
fig = plt.figure(figsize=(3,2),dpi=100)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2,figsize=(8,4),dpi=100)
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
x
```

```python
z
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('X --->')
ax.set_ylabel('Y --->')
ax.set_title('Title')
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.2,.2])
ax1.plot(x,y,color='red')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax2.plot(x,y,color='red')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax1.set_xlim([0,100])
ax1.set_ylim([0,10000])
ax1.plot(x,z)
ax1.set_xlabel('x')
ax1.set_ylabel('z')
ax2 = fig.add_axes([0.2,0.5,.4,.4])
ax2.set_xlim([20,22])
ax2.set_ylim([30,50])
ax2.plot(x,y)
ax2.set_title('zoom')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y,color='blue',ls='--',lw=3)
axes[1].plot(x,z,color='red',lw=3)
plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2,figsize=(10,2))
axes[0].plot(x,y,color='blue',ls='--',lw=3)
axes[1].plot(x,z,color='red',lw=3)
plt.tight_layout()
```
