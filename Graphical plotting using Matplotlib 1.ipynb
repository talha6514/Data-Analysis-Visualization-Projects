```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
x
```

```python
fig,axes = plt.subplots(1,2,figsize=(10,4))
axes[0].plot(x,x**2,x,np.exp(x)) #In the format (x1,y1,x2,y2)
axes[0].set_title("Normal Scale")
axes[1].plot(x,x**2,x,np.exp(x))
axes[1].set_yscale("log")
axes[1].set_title('Logarithmic Scale(y)')
```

```python
y = 2
print(np.exp(y))
```

```python
fig,ax = plt.subplots(figsize=(10,4))
ax.plot(x,x**2,x,x**3,lw=2)

ax.set_xticks([1,2,3,4,5])
ax.set_xticklabels([r'$\alpha$',r'$\beta$',r'$\gamma$',r'$\delta$', r'$\epsilon$'],fontsize=18)

yticks = [0,50,100,150]
ax.set_yticks(yticks)
ax.set_yticklabels(["$%.1f$" % y for y in yticks],fontsize=18)
```

```python
fig,ax = plt.subplots(1,1)
ax.plot(x,x**2,x,np.exp(x))
ax.set_title("Scientific Notation")
ax.set_yticks([0,50,100,150])

from matplotlib import ticker
formatter = ticker.ScalarFormatter(useMathText=True)
formatter.set_scientific(True)
formatter.set_powerlimits((-1,1))
ax.yaxis.set_major_formatter(formatter)
```

```python
mpl.rcParams['xtick.major.pad'] = 5 #matplotlib.rcParams
mpl.rcParams['ytick.major.pad'] = 5
fig,ax = plt.subplots(1,1)
ax.plot(x,x**2,x,np.exp(x))
ax.set_yticks([0,50,100,150])
ax.set_title('Lable & Axis Spacing')

# padding between axis label and sxis numbers
ax.xaxis.labelpad = 5
ax.yaxis.labelpad = 5
ax.set_xlabel('X')
ax.set_ylabel('Y');
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10,3))

# default grid appearance
axes[0].plot(x, x**2, x, x**3, lw=2)
axes[0].grid(True,linewidth=2)

# custom grid appearance
axes[1].plot(x, x**2, x, x**3, lw=2)
axes[1].grid(color='b', alpha=0.5, linestyle='dashed', linewidth=1)
```

```python
fig,ax = plt.subplots(figsize=(6,2))
ax.spines['bottom'].set_color('blue')
ax.spines['top'].set_color('blue')
ax.spines['left'].set_color('red')
ax.spines['left'].set_linewidth(2)
# turnoff axis spin to the right
ax.spines['right'].set_color('none')
ax.yaxis.tick_left()
```

```python
fig, ax1 = plt.subplots()

ax1.plot(x, x**2, lw=2, color="blue")
ax1.set_ylabel(r"area $(m^2)$", fontsize=18, color="blue")
for label in ax1.get_yticklabels():
    label.set_color("blue")
    
ax2 = ax1.twinx()
ax2.plot(x, x**3, lw=2, color="red")
ax2.set_ylabel(r"volume $(m^3)$", fontsize=18, color="red")
for label in ax2.get_yticklabels():
    label.set_color("red")
```

```python
fig,axes = plt.subplots()
axes.spines['top'].set_color('None')
axes.spines['right'].set_color('None')

axes.xaxis.set_ticks_position('bottom')
axes.spines['bottom'].set_position(('data',0))

axes.yaxis.set_ticks_position('left')
axes.spines['left'].set_position(('data',0))

xx = np.linspace(-0.75,1.0,100)
axes.plot(xx,xx**3)
```

# Some 2D Plots

```python
n = np.array([1,2,3,4,5])
```

```python
fig,axes = plt.subplots(1,4,figsize=(12,3))
axes[0].scatter(xx, xx + 0.25*np.random.randn(len(xx)))
axes[0].set_title("Scatter")

axes[1].step(n,n**2,lw=2,color='Green')
axes[1].set_title('Step')

axes[2].bar(n,n**2,align='center',width=0.5)
axes[2].set_title('Bar')

axes[3].fill_between(x,x**2,x**3,color='green')
axes[3].set_title('Fill_Between')
plt.tight_layout()
```

# Figures with multiple subplots and insets

Axes can be added to a matplotlib Figure canvas manually using 'fig.add_axes' or using a sub-figure layout manager such as 'subplots', 'subplot2grid', or 'gridspec':

```python
fig = plt.figure()
ax1 = plt.subplot2grid((3,3), (0,0), colspan=3)
ax2 = plt.subplot2grid((3,3), (1,0), colspan=2)
ax3 = plt.subplot2grid((3,3), (1,2), rowspan=2)
ax4 = plt.subplot2grid((3,3), (2,0))
ax5 = plt.subplot2grid((3,3), (2,1))
fig.tight_layout()
# Reference link - https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.subplot2grid.html
```

```python
import matplotlib.gridspec as gridspec
```

```python
fig = plt.figure()

gs = gridspec.GridSpec(2,3,height_ratios=[2,1],width_ratios=[1,2,1])
for g in gs:
    ax = fig.add_subplot(g)
    
fig.tight_layout()
```

## add_axes

```python
fig, ax = plt.subplots()

ax.plot(xx, xx**2, xx, xx**3)
fig.tight_layout()

# inset
inset_ax = fig.add_axes([0.25,0.55, 0.35, 0.35]) # X, Y, width, height

inset_ax.plot(xx, xx**2, xx, xx**3)
inset_ax.set_title('zoom near origin')

# set axis range
inset_ax.set_xlim(-.2, .2)
inset_ax.set_ylim(-.005, .01)

# set axis tick locations
inset_ax.set_yticks([0, 0.005, 0.01])
inset_ax.set_xticks([-0.1,0,.1]);
```
